```python
def problem_1(message, shift):
    word_one = ''
    for i in message:
        if ord(i) >= 65 and ord(i) <= 90:
            i = chr((ord(i) + shift - 65) % 26 + 65)
            if ord(i) == 32:
                None
        word_one = word_one + i

        
    return(word_one)

social_graph = {
    "@bongolpoc":{"first_name":"Joselito",
                  "last_name":"Olpoc",
                  "following":[
                  ]
    },
    "@joaquin":  {"first_name":"Joaquin",
                  "last_name":"Gonzales",
                  "following":[
                      "@chums","@jobenilagan"
                  ]
    },
    "@chums" : {"first_name":"Matthew",
                "last_name":"Uy",
                "following":[
                    "@bongolpoc","@miketan","@rudyang","@joeilagan"
                ]
    },
    "@jobenilagan":{"first_name":"Joben",
                   "last_name":"Ilagan",
                   "following":[
                    "@eeebeee","@joeilagan","@chums","@joaquin"
                   ]
    },
    "@joeilagan":{"first_name":"Joe",
                  "last_name":"Ilagan",
                  "following":[
                    "@eeebeee","@jobenilagan","@chums"
                  ]
    },
    "@eeebeee":  {"first_name":"Elizabeth",
                  "last_name":"Ilagan",
                  "following":[
                    "@jobenilagan","@joeilagan"
                  ]
    },
}

def relationship_status(a_member,b_member,social_graph):
    
    stat = ""   
    if b_member in social_graph[a_member]["following"] and a_member in social_graph[b_member]["following"]:
        stat = "Friends"
        
    elif b_member in social_graph[a_member]["following"]:
        stat = "Follower"
        
    elif a_member in social_graph[b_member]["following"]:
        stat = "Followed by"
    else:
        stat = "None"
        
    print(stat)
a_member = input()
b_member = input()
stat = relationship_status(a_member,b_member,social_graph)

def tic_tac_toe(board):
         
    wins = ""    
    
    board_size = len(board)
    
    while wins == "":

        #for rows
        for row in board:
            if all(move == "X" for move in row):
                wins = "X"
            elif all(move == "O" for move in row):
                wins = "O"
        if wins == "X" or wins == "O":
            break
            
        #for columns
        for h in range(board_size):
            if all(move == "X" for move in (list(board[i][h] for i in range(board_size)))):
                wins = "X"
            elif all(move == "O" for move in (list(board[i][h] for i in range(board_size)))):
                wins = "O"
        if wins == "X" or wins == "O":
            break
        
        #for diagonals
        if all(move == "X" for move in (list(board[i][i] for i in range(board_size)))) or all(move == "X" for move in (list(board[board_size - 1 - i][i] for i in range(board_size)))):
            wins = "X"
        elif all(move == "O" for move in (list(board[i][i] for i in range(board_size)))) or all(move == "O" for move in (list(board[board_size - 1 - i][i] for i in range(board_size)))):
            wins = "O"
        if wins == "X" or wins == "O":
            break
            
       
        wins = "None"
    
    print(wins)
    
board1 = [
['X','O','X'],
['X','X','O'],
['X','','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','','X'],
['O','X','O'],
['O','O','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','O',''],
]

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

def eta(legs,start,end):
    time = 0
    i = True
    
    while i == True:
        for key in legs:
            cleg = key[0]
            nleg = key[1]
            tt = legs[key]["travel_time_mins"]
            
            if cleg == start:
                start = nleg
                time += tt
            if start == end:
                i = False
                break
                
    print(time)
start = input()
end = input()

time = eta(legs,start,end)
```
